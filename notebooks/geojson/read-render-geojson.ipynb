```python
# Import necessary library
import folium
import geopandas as gpd
```

```python
# Function to create a map and render GeoJSON
def render_geojson(m, geojson_path, color='blue', opacity=0.5, fill_opacity=0.5):
    # Add the GeoJSON to the map
    folium.GeoJson(
        geojson_path,
        style_function=lambda x: {
            'fillColor': color,
            'color': color,
            'weight': 2,
            'opacity': opacity,
            'fillOpacity': fill_opacity
        },
        name='geojson'
    ).add_to(m)
    
    # Display the map
    return m

def initialize_map(geojson_path, zoom_start=7):
    # Read the GeoJSON file using GeoPandas
    gdf = gpd.read_file(geojson_path)
    
    # Get the bounds of the GeoDataFrame
    bounds = gdf.total_bounds  # returns [minx, miny, maxx, maxy]
    
    # Calculate the center based on the bounds
    center = [(bounds[1] + bounds[3]) / 2, (bounds[0] + bounds[2]) / 2]
    
    # Initialize the Folium map
    m = folium.Map(location=center, zoom_start=zoom_start)
    
    return m
```
