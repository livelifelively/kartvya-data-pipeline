```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
triplets = [
    ("Politics of India", "works within", "Constition of India"),
    ("India", "government_system", "parliamentary"),
    ("India", "government_system", "republic"),
    ("India", "government_system", "secular"),
    ("India", "government_system", "democratic"),
    ("India", "head_of_state", "President of India"),
    ("President of India", "is_a", "First citizen of India"),
    ("President of India", "is_a", "Head of government"),
    
    ("President of India", "is_a", "First citizen of India"),
]
```

```python
graph = nx.DiGraph()
```

```python
for subject, predicate, object in triplets:
    graph.add_node(subject)
    graph.add_node(object)
    graph.add_edge(subject, object, label=predicate)# Draw the graph
pos = nx.spring_layout(graph)  # Node positioning algorithm
nx.draw(graph, pos, with_labels=True, node_size=500, node_color="skyblue")
edge_labels = nx.get_edge_attributes(graph, 'label')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
plt.title("Graph Visualization")
plt.show()
```
